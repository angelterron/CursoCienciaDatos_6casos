# TAREA #8: LLEVAR A CABO LA LIMPIEZA DE DATOS APLICANDO TODO LO QUE HEMOS APRENDIDO!!

```python
# Definimos un pipeline para limpiar todos los mensajes
# El pipeline realizamos lo siguiente: (1) eliminar signos de putuaci칩n, (2) eliminar las stopwords
def message_cleaning(message):
    message_remove = [char for char in message if char not in string.punctuation]
    message_join = "".join(message_remove)
    message_clean = [word for word in message_join.split() if word.lower() not in stopwords.words('english')]
    return message_clean
```

```python
# Probemos la funci칩n que acabamos de crear

reviews_df_clean = reviews_df['verified_reviews'].apply(message_cleaning)
```

# TAREA #10: ENTRENAR UN MODELO CLASIFICADOR DE NA칊VE BAYES

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

nb_multinomial = MultinomialNB()

nb_multinomial.fit(X_train, y_train)

```

```python
nb_multinomial
```

# TAREA #11: VALIDAR LA EFICACIA DEL MODELO ENTRENADO

![alt text](https://drive.google.com/uc?id=1ZKbO9tQH3qh0IjKwgOAJ37ghWzUVcb8O)

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_test = nb_multinomial.predict(X_test)

cm = confusion_matrix(y_test, y_predict_test)

sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict_test))
```

# TAREA #12: RETO FINAL - ENTRENAR Y EVALUAR UN MODELO CLASIFICADOR BASADO EN LA REGRESI칍N LOG칈STICA

```python
y_predict_test = model.predict(X_test)

cm = confusion_matrix(y_test, y_predict_test)

sns.heatmap(cm, annot = True)
```
